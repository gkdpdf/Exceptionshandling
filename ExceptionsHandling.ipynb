5. Custom exception is used to give customised messages and handle customised errors
For example age can't be negative to handle it separtely we can raise our own customised exceptions


```python
#6# define Python user-defined exceptions
class InvalidAgeException(Exception):
    def __init__(self,msg):
        self.msg=msg
    

# you need to guess this number
number = 18

try:
    input_num = int(input("Enter a number: "))
    if input_num < number:
        raise InvalidAgeException( "Raised when the input value is less than 18")
    else:
        print("Eligible to Vote")
        
except InvalidAgeException:
    print("Exception occurred: Invalid Age")
```
